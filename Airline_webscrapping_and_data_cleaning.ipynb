```python
# Function to remove punctuation, lowercase all letters and perform tokenization
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\d+','', text) #removing numbers
    text = re.sub(r'[^\w\s]','', text) #removing puntuation and special character
    
    tokens = nltk.word_tokenize(text) # split text into individual tokens
    
    return tokens
```

```python
# Function to remove stop words
def remove_stopwords(tokens):
    stop_words = set(stopwords.words('english')) #importing stop words
    filtered_tokens = [word for word in tokens if word not in stop_words]
    return filtered_tokens
```

```python
# Function to reduce words to their base format...The base form is a standard dictionary form of a word (e.g., "running" -> "run").
def perform_lemmatization(tokens):
    lemmatizer = nltk.WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return lemmatized_tokens
```

```python
# Function to apply previous functions to the selected text
def clean_text(text):
    tokens = preprocess_text(text)
    filtered_tokens = remove_stopwords(tokens)
    lemmatized_tokens = perform_lemmatization(filtered_tokens)
    clean_text = ' '.join(lemmatized_tokens)
    return clean_text

# Apply the clean_text function to the entire review data.
df['reviews'] = df['reviews'].apply(clean_text)
```

```python
# convert the date to datetime format
def clean_and_convert_date(date_str):
    # Remove ordinal suffixes
    cleaned_date_str = re.sub(r'(\d+)(st|nd|rd|th)',r'\1', date_str)
    # Convert to datetime object
    datetime_obj = pd.to_datetime(cleaned_date_str, format="%d %B %Y")
    # Format to "YYYY-MM-DD"
    formatted_date = datetime_obj.strftime("%Y-%m-%d")
    return formatted_date
df['date'] = df['date'].apply(clean_and_convert_date) 
```
